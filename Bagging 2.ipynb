```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_wine
```

```python
wine=load_wine()
```

```python
df=np.c_[wine.data,wine.target]
```

```python
df=pd.DataFrame(df,columns=wine.feature_names+['target'])
```

```python
df
```

```python
df['target'].value_counts
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,13]
```

```python
from sklearn.model_selection import cross_val_score
```

```python
s=cross_val_score(dt,x,y,cv=10)
```

```python
print(np.round(np.mean(s),2))
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bg=BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    max_samples=0.5,
    bootstrap=True,
    n_estimators=500,
    random_state=42
)
```

```python
e=cross_val_score(bg,x,y,cv=10)
```

```python
print(np.round(np.mean(e),2))
```
